```python
# Concatenate BLAST outputs
!cat blast/all_bac.fasta blast/excl_coli.fasta > blast/combined_blast.fasta
```

```python
# CD-HIT to reduce redundancy
!cd-hit \
    -i blast/combined_blast.fasta \
    -o cdhit/combined_cd.fasta \
    -c 0.985
```

```python
# Multiple Sequence Alignment
!linsi cdhit/combined_cd.fasta > mafft/combined.mafft.fasta
```

```python
# Clean alignment
!trimal -in mafft/combined.mafft.fasta -out trimal/trimal.auto1.fasta -automated1
```

```python
# Test which model to use
!modeltest-ng -i trimal/trimal.auto1.fasta -d aa -p 8 -T raxml -o modeltest/modeltestaa
```

```python
# Generate Maximum-Likelihood phylogenetic tree
!raxml-ng --all \
--msa trimal/trimal.auto1.fasta \
--model FLU+G4 \
--prefix raxml/newtree_aa2 \
--bs-trees 1000 \
--threads 8 \
--seed 2
    
```
