# Trimal on MAFFT allignment

There were a lot of truncated sequences, so here I use trimAL to automatically trim the allignments.

The default parameters did not remove the truncated sequences effectively. This was likely due to the fact most of the gaps in sequences were at the start or end. 

I went through a number of iterations, manually editing the parameters until I got a satisfactory, well-trimmed output. The parameters I chose (as shown below) resulted in only remaining truncated sequences which I manually removed in Jalview. I also manually removed two adjacent empty columns which were filled by only 1 sequence.

The parameters `-resoverlap 0.7 -seqoverlap 80` were the most effective.

```python
# trimAL parameters
trimal -in ../../data/cleaned/espy3-50.mafft.fasta -out ../../data/cleaned/trimal/espy3-50.mafft.r07s08.trim.fasta -keepheader -resoverlap 0.6 -seqoverlap 80
```
