# Uniprot 50% Sequence Identity MAFFT Alignment - EspY3

**2025-10-30**


From the EspY3 Uniprot page (Accession: Q8XE85) I retrieved the EspY3 homologs at 50% sequence identity and alligned with MAFFT

I chose L-INS-i allignment strategy as is recommened by the MAFFT manual for best allignment <200 sequences.

**Input:**: `espy3-50.fasta`
**Output:** `espy3-50.mafft.fasta`



```python
mafft --localpair --maxiterate 16 --reorder ../../data/raw/espy3-50.fasta > ../../data/cleaned/espy3-50.mafft.fasta
```
